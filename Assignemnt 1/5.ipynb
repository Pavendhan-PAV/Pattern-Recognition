```python
# importing essential libraries
import cv2
import numpy
import math
from matplotlib import pyplot as plt
```

```python
# read image files
query_img = cv2.imread('Query_image.jpg',0)   
ref_1 = cv2.imread('Reference_image1.jpg',0)   
ref_2 = cv2.imread('Reference_image2.jpg',0) 
```

```python
# calculate histogram of image and flatten lists
query_hist = list(numpy.concatenate(cv2.calcHist([query_img],[0],None,[256],[0,256])).flat)
ref_1_hist = list(numpy.concatenate(cv2.calcHist([ref_1],[0],None,[256],[0,256]) ).flat)
ref_2_hist = list(numpy.concatenate(cv2.calcHist([ref_2],[0],None,[256],[0,256]) ).flat)
```

### Histogram Plotting

```python
# plotting histogram of query image 
plt.plot(query_hist)
plt.show()
plt.imshow(cv2.cvtColor(query_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# plotting histogram of reference image 1
plt.plot(ref_1_hist)
plt.show()
plt.imshow(cv2.cvtColor(ref_1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# plotting histogram of reference image 2
plt.plot(ref_2_hist)
plt.show()
plt.imshow(cv2.cvtColor(ref_2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# normalize lists
norm = max(query_hist) - min(query_hist)
query_hist = [i/norm for i in query_hist]
norm = max(ref_1_hist) - min(ref_1_hist)
ref_1_hist = [i/norm for i in ref_1_hist]
norm = max(ref_2_hist) - min(ref_2_hist)
ref_2_hist = [i/norm for i in ref_2_hist]
```

### Euclidean Distance

```python
# calculate euclidean distance between query image and reference image 1
euc_dist = 0.0
for i in range(len(query_hist)):
    euc_dist += (query_hist[i] - ref_1_hist[i]) ** 2
print("Euclidean distance between query image and reference image 1: ",math.sqrt(euc_dist))
```

```python
# calculate euclidean distance between query image and reference image 2
euc_dist = 0.0
for i in range(len(query_hist)):
    euc_dist += (query_hist[i] - ref_2_hist[i]) ** 2
print("Euclidean distance between query image and reference image 2: ",math.sqrt(euc_dist))
```
